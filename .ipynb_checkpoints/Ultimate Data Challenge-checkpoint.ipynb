```python
# Data processing for mahcine learing part

# One hot encode the categorical city variables
df_ohe = pd.get_dummies(df2, columns=['city', 'phone', 'ultimate_black_user'])

# We will drop the datetime as this is not needed for training and testing (and won't pass in our models)
df_ohe = df_ohe.drop(['signup_date', 'last_trip_date'], axis=1)

# Create separate object for target variable
y = df_ohe['stayed_3mo']

# Create separate object for input features
X = df_ohe.drop(['stayed_3mo', 'stayed_6mo'], axis=1)
X
```

```python
# Split-out validation dataset using stratified random sampling


# Split X and y into train and test sets
X_train, X_test, y_train, y_test  = train_test_split(X, y,
                                                     test_size = 0.2,
                                                     random_state = 1234, 
                                                     stratify = df_ohe.stayed_3mo)

# Print number of observations in X_train, X_test, y_train, and y_test
print(len(X_train), len(X_test))
print(len(y_train), len(y_test))
```

We'll create Pipelines (instead of GridSearch() since I haven't used it in any of my projects so let's try). The purpose of the pipeline is to assemble several steps that can be cross-validated together while setting different parameters.

```python
# Pipeline dictionary
pipelines = {
    'lr': make_pipeline(StandardScaler(),
                        LogisticRegression(random_state=123)),
    'rf': make_pipeline(StandardScaler(),
                        RandomForestClassifier(random_state=123)),
    'gb': make_pipeline(StandardScaler(),
                        GradientBoostingClassifier(random_state=123))
}
```

```python
# Hyperparameter tuning

# List tuneable hyperparameters of our pipeline
pipelines['lr'].get_params()
```

```python
# # Logistic Regression hyperparameters
lr_hyperparameters = {
    'logisticregression__C': np.logspace(-4, 4, 20),
    'logisticregression__solver': ['lbfgs','newton-cg','liblinear','sag','saga']
}
```

```python
# Random Forest hyperparameters
rf_hyperparameters = {
    'randomforestclassifier__n_estimators': [100,200],
    'randomforestclassifier__max_features': ['auto','sqrt',0.33]
}

```

```python
# Boosted Tree hyperparameters
gb_hyperparameters = {
    'gradientboostingclassifier__n_estimators': [100,200],
    'gradientboostingclassifier__learning_rate': [0.05, 0.1,0.2],
    'gradientboostingclassifier__max_depth': [1,3,5]
}

```

```python
# Create hyperparameters dictionary
hyperparameters = {'lr': lr_hyperparameters,
                  'rf': rf_hyperparameters,
                  'gb': gb_hyperparameters}
```

### Fitting the models to the train set

```python
# Create empty dictionary called fitted_models
fitted_models = {}

# Loop through model pipelines, tuning each one and saving it to fitted_models
for name, pipeline in pipelines.items():
    # Create cross-validation object from pipeline and hyperparameters
    model = GridSearchCV(pipeline, hyperparameters[name], cv = 10, n_jobs=-1)
    
    # Fit model on X_train, y_train
    model.fit(X_train, y_train)
    
    # Store model in fitted_models[name] 
    fitted_models[name] = model
    
    # Print '{name} has been fitted'
    print('{0} has been fitted'.format(name))
```
